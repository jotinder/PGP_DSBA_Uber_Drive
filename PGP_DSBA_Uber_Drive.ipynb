## Q22. What is proportion of trips that is Business and what is the proportion of trips that is Personal? (3 points)
### Note: The proportion calculation  is with respect to the 'miles' variable.
### Note: This question is based on the original dataframe.

```python
print(color.RED + color.BOLD + color.UNDERLINE + 'Propotion of trips that is business and Propotion of trips that is Personal (By Miles)' + color.END)
display((uber_df.groupby('CATEGORY*').sum()) / (uber_df.groupby('CATEGORY*').sum().sum()))
print('\n')
print(color.RED + color.BOLD + color.UNDERLINE + 'Percent of trips that is business and Percent of trips that is Personal (By Miles)' + color.END)
display((uber_df.groupby('CATEGORY*').sum()) / (uber_df.groupby('CATEGORY*').sum().sum()) * 100)
```
